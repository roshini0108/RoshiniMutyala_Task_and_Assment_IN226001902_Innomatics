Problem 1: Mobile Recharge Validation System

```python
def recharge(t):
    if t in [199,299,399,599]:
        return "recharge plan successful"
    else:
        return "retry payment"
while True:
    t=int(input("enter recharge plan amount: "))
    r=recharge(t)

    if r=="recharge plan successful":
        print(r)
        break
    else:
        print("Invalid plan. Please try again.")
```

Problem 2: Inventory Reorder Alert System

```python
def check_inventory(products):
    for item,quantity in products.items():
        if quantity<15:
            print(item, ":", "Reorder Alert")
        else:
            print(item, ":", "Stock OK")
inventory = {}
n = int(input("Enter number of products: "))
for i in range(n):
    product=input("Enter product name: ")
    quantity=int(input("Enter stock quantity: "))
    inventory[product]=quantity
check_inventory(inventory)
```

 Problem 3: Student Result Processing System

```python
def marks(name,rollno):
  n=int(input("enter number of subjects: "))
  l=list(map(int,input("enter marks: ").split()))
  tot=0
  for i in l:
    tot+=i
  if tot/n>=50:
    return "pass"
  else:
    return "fail"
name=input("enter name: ")
rollno=input("enter rollno: ")
print(f"{name} of roll no {rollno} is: ",marks(name,rollno))
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def cab_fare():
    distance=int(input("enter distance in km: "))
    p1km=distance*12 #as 12 rupees price for 1km
    time=int(input("enter time you want to travel in: "))
    if distance<=0 or time<=0 or time>24:
      return "invalid"
    #lets consider peak hours as mrng 9-11am and evng 5-8pm
    if time in range(9,12) or time in range(17,20):
      return p1km+p1km*(0.25)
    else:
      return p1km
while True:
    result = cab_fare()
    if result == "invalid":
        print("Invalid input! Please retry.\n")
        continue
    else:
        print("Total Fare: ₹", result)
        break
```

Problem 5: Employee Attendance Eligibility Checker

```python
def attendance(name,rollno):
  n=int(input("enter number of days: "))
  l=input("Enter P/A: ").upper().split()
  if any(v not in ['P', 'A'] for v in l):
      return "invalid"
  cp=0
  for i in l:
    if i=="P":
      cp+=1
  if cp/n>=0.75:
    return "Elgibile"
  else:
    return "Not Elgibile"
name=input("enter name: ")
rollno=input("enter rollno: ")
print(f"{name} of roll no {rollno} is:",attendance(name,rollno),"to write exams")
```

Problem 6: Password Strength Checker

```python
def check_password(password):
    if len(password)<8:
        return "Weak Password (Minimum 8 characters required)"
    has_digit=False
    has_special=False
    for ch in password:
        if ch.isdigit():
            has_digit=True
        if ch in "@#$":
            has_special=True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character @#$)"
password = input("Enter your password: ")
result = check_password(password)
print(result)
```
