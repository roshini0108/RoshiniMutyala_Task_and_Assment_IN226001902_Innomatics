2. Pass / Fail Analyzer :

```python
'''enter the student marks to found out no of students pass or fail'''
marks=list(map(int,input("Enter Marks").split()))
count_pass=0
count_fail=0
for i in marks: # iterate through all marks
  if i>=50: #if marks above 50 student pass
    count_pass+=1
  else:
    count_fail+=1 # else fail
print("Number of students passed:",count_pass)
print("Number of students failed:",count_fail)
```

3. Simple Data Cleaner

```python
'''enter names'''
names=[" Alice ", "bob", " CHARLIE "]
cleaned_list=[]
for name in names:
  cleaned_names=name.strip().lower() #strip remove white spaces and lower converts all letters to lower case
  cleaned_list.append(cleaned_names)
print(cleaned_list)
```

4. Message Length Analyzer

```python
messages=list(map(str,input("Enter messages:").split()))
for message in messages:
  print(len(message)) #prints length of string
  if len(message)>10: # if len becomes more then 10 it prints 10 and the flag message
    print(" FLAG: message length greater then 10 is not acceptable")
```
